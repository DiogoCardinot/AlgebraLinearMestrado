5- Crie um array random (com elementos aleatórios inteiros) entre 0 e 10 de tamanho (7, 2)

6- Encontre os elementos mínimo e máximo e calcule a soma, a média e o desvio padrão

```python
C.min()
C.max()
C.sum()
C.mean()
C.std()
print("Mínimo: ", C.min())
print("Máximo: ", C.max())
print("Soma: ", C.sum())
print("Média: ", C.mean())
print("Desvio Padrão: ", C.std())
```

8- Crie duas listas, uma chamada x, outra chamada y cada uma com 5 números

 Plote as listas x e y separadamente

```python
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
plt.plot(x, label=f'x')
plt.plot(y, label=r'y')
plt.legend()
plt.xlabel(r'Valores de x')
plt.ylabel(r'Valores de y')
```

10- Crie um plot usando plt.subplots() e plot x e x² (squared)

```python
def f1(x):
    return x
def f2(x):
    return x**2

t1 = np.random.randint(1,50,size=(1,5))
t2 = np.random.randint(1,50,size=(1,5))

plt.subplot(211)
plt.plot(t1, f1(t1),'o')
plt.subplot(212)
plt.plot(t2, f2(t2), 'bo')
```
