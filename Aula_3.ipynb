## **Exemplo 1**

$\left\{\array{x+y+z=1\\x+2y+2z=2\\x+4y+5z=4}\right.$

$\left[\array{1&1&1\\1&2&2\\1&4&5}\right]$$\left[\array{x\\y\\z}\right]$=$\left[\array{1\\2\\4}\right]$

```python
A1 = np.array([[1, 1, 1],
              [1, 2, 2],
              [1, 4, 5]])
```

```python
np.linalg.solve(A1,B1)
```

## **Exemplo 1**

$\left[\array{1&1\\-1&1}\right]$$\left[\array{x\\y}\right]$=$\left[\array{1\\5}\right]$

```python
# definindo uma matriz matrizA
matrizA = np.array([[1,1],
                   [-1,1]])
```

```python
matrizA
```

```python
 # definindo uma matriz matrizB
matrizB = np.array([[1],[5]])
```

### **Posto da Matriz de Coeficientes**

```python
np.linalg.matrix_rank(matrizA)
```

### **Posto da Matriz Ampliada**

```python
matrizAA = np.array([[1,1,1],
                    [-1,1,5]])
```

```python
np.linalg.matrix_rank(matrizAA)
```

Um sistema de m equações e n incógnitas admite solução se, e
somente se, o posto da matriz ampliada é igual ao posto da
matriz dos coeficientes

Se as duas matrizes têm o mesmo posto p e p = n (número de
incógnitas), então a solução será única

$y=1−x$

$y=5+x$

```python
x,y=np.linalg.solve(matrizA,matrizB)
```

```python
x
```

```python
y
```

```python
# define valores para  a variável x, entre -4 e 2
valoresX = np.linspace(-4,2)
```

```python
valoresX
```

```python
# coeficientes associadas ao x
x1, x2 = -1, 1
```

```python
# valor resultante da operação do sistema
b1, b2 = 1, 5
```

```python
# definindo o tamanho da figura
plt.figure(figsize=(10,5))
```

```python
# plota a primeira reta y = 1 - x
plt.plot(valoresX, x1 * valoresX + b1, color = "Red", label = "y = 1 - x ")
# plota a segunda reta y = 5 + x
plt.plot(valoresX, x2 * valoresX + b2, color = "Yellow", label = " y = 5 + x")
# adiciona o ponto de interseção entre as retas
plt.scatter(x, y, color = "Green", label = "Valores para (x,y)")
# adiciona legendas ao gráfico
plt.legend()
# remove a caixa preta nas bordas do gráfico
plt.box(False)
# desenho do quadriculado do gráfico
plt.grid(True)
plt.title("Gráfico que relaciona y em função de x")
```

## **Exemplo 2**

```python
matrizC = np.array([[4,7],[2,3.5]])
```

```python
matrizC
```

### **Posto Matriz de Coeficientes**

```python
np.linalg.matrix_rank(matrizC)
```

### **Posto Matriz Ampliada**

```python
matrizCA = np.array([[4,7,1],
                    [2,3.5,5]])
```

```python
np.linalg.matrix_rank(matrizCA)
```

Posto da matriz de coeficientes e da matriz ampliada são diferentes logo não tem solução

$y=\frac{−4x+1}{7}$

$y=\frac{−2x+5}{3.5}$

```python
# define valores para a variável x, entre -4 e 2
valoresX = np.linspace(-4,2)
```

```python
# coordenadas associadas a x
x1, x2 = -4 / 7, -2 / 3.5
```

```python
# valor resultante da operação do sistema
b1, b2 = 1 / 7, 5 / 3.5
```

```python
# definindo o tamanho da figura
plt.figure(figsize = (10,5))
```

```python
# plota a primeira reta y = (-4x + 1) / 7
plt.plot(valoresX, x1 * valoresX + b1, color = "Red", label = "y = (-4x + 1) / 7 ")
# plota a segunda reta y = (-2x + 5) / 3.5
plt.plot(valoresX, x2 * valoresX + b2, color = "Yellow", label = " y = (-2x + 5) / 3.5")
# adiciona legendas ao gráfico
plt.legend()
# remove a caixa preta nas bordas do gráfico
plt.box(False)
# desenho do quadriculado do gráfico
plt.grid(True)
plt.title("Gráfico que relaciona y em função de x")
```

## **Exemplo 3**

$\left[\array{1&2&7\\0&1&3\\-3&1&0}\right]$$\left[\array{x\\y\\z}\right]$=$\left[\array{0\\0\\0}\right]$

```python
# definindo uma matriz
matrizD = np.array([[1,2,7],
                   [0,1,3],
                   [-3,1,0]])
```

```python
matrizD
```

```python
matrizE = [0,0,0]
```

### **Posto da Matriz de Coeficientes**

```python
np.linalg.matrix_rank(matrizD)
```

### **Posto da Matriz Ampliada**

```python
matrizDA = np.array([[1,2,7,0],[0,1,3,0],[-3,1,0,0]])
```

```python
np.linalg.matrix_rank(matrizDA)
```

Um sistema de m equações e n incógnitas admite solução se, e somente se, o posto da matriz ampliada é igual ao posto da matriz dos coeficientes

Se as duas matrizes têm o mesmo posto p e p < n, então podemos escolher n – p incógnitas, as outras p incógnitas serão dadas em função dessas

2 < 3

3-2 = 1 variável livre

Vamos resolver metodicamente e descobrir o valor de cada componente do sistema de equações:

Pela $L_3$

$ y = 3x$

Pela $L_2$

$ y = -3z$

Logo

$ 3x = - 3z$

Assim, teremos:

$ x = -z$

Assim, veja que z é uma variável livre.

```python
# define valores arbitrários para a variável z, entre -4 e 2
valoresZ = np.linspace(-4,2)
```

```python
# coordenadas associadas a x e y
x, y = -1, -3
```

```python
ax = plt.figure().add_subplot(projection='3d')

# Plot a sin curve using the x and z axes.

ax.plot(valoresZ, x*valoresZ, zs=0, zdir='y', label='x=-z')
ax.plot(valoresZ, y*valoresZ, zs=0, zdir='y', color ='red',label='y=-3z')
plt.legend()
```

```python
np.linalg.solve(matrizD,matrizE)
```
